```python
# Local imports
import torch
from src import utils
from src.hmm import HMM
from transformers import AutoTokenizer

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
hmm_model_path = 'models/hmm_gpt2-large_uncon_seq-len-32_4096_10M'
model_path = 'gpt2-large'
weights_path = 'data/coefficients.csv'

# Load HMM model
# hmm_model = HMM(hidden_size=4096, vocab_size=50000, eos_token_id=50256, sep_token_id=50256)
hmm_model_rand = HMM(hidden_size=4096, vocab_size=50257, eos_token_id=50256)
hmm_model: HMM = utils.load_hmm_model(hmm_model_path, device=device)
weights_tensor = utils.load_weights(weights_path, device=device)
hmm_model.set_weights(weights_tensor)
# print hmm attrs:
for attr in ['hidden_size', 'vocab_size', 'eos_token_id', 'sep_token_id']:
    print(f"{attr}: {getattr(hmm_model, attr)}")

# Load tokenizer
gen_tokenizer = AutoTokenizer.from_pretrained(model_path, padding_side="left")
gen_tokenizer.pad_token = gen_tokenizer.pad_token or gen_tokenizer.eos_token
```
